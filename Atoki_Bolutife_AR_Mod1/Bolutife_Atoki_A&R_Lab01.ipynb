**INTERPRETATIONS**

1a) A phantom is a specially designed object or material used for calibration, quality control, and testing of CT imaging systems. Phantoms are used to simulate the properties of human tissues or other materials of interest in order to assess the performance and accuracy of the CT scanner.

1b) What makes the Shepp Logan Phantom special is its mathematical simplicity and its ability to mimic the characteristics of human anatomy, which makes it a valuable tool for testing and evaluating image reconstruction algorithms and CT systems. Also since it is generated and customized using a well-defined mathematical model, the ground truth of the phantom is known precisely, meaning that researchers can compare the reconstructed images produced by a CT scanner or image reconstruction algorithm with the known original phantom, allowing for quantitative assessment of the accuracy and quality of the reconstruction

2a) A sinogram is a graphical representation X-ray projectstions of an object, acquired during a computed tomography (CT) scan by sending rays through the object and having a receiver collect the rays that pass through, in such a way that this transmitter-receiver system is revolved around the object. It is obtained using mathematical transformations like the Radon transform.

2b) Upon inspecting the [source code](https://github.com/scikit-image/scikit-image/blob/main/skimage/transform/radon_transform.py) of skimage's radon function, it was observed that the rotation direction is determined by the circle parameter, which specifies the center of rotation for the Radon transform, which is set to False (Counter-clockwise) by default.

4) The Radon transform is a linear operator meaning that the radon transform of the sum of two functions f and g is the same as the sum of their individual radon transforms.

**Image Reconstruction from Sinogram using inverse radon transform**

**INTERPRETATION**

The default filter for skimage's inverse radon transform is the Ramp filter.

The ramp filter helps to correct the blurring effect that occurs during data acquisition. The filter has a frequency response that is proportional to the frequency of the sinusoidal components present in the data, meaning  that it enhances high-frequency components (i.e., fine details) and attenuates low-frequency components (i.e., low-contrast structures and artifacts), resulting in a sharpened image.

```python
#varying the number of projections
projections = [50, 100, 500, 1000] #number of projections to consider

plt.figure(figsize=(12, 12))

for i, projection in enumerate(projections):

  theta = np.linspace(0., ANGLE, projection, endpoint=False)
  sinogram = radon(image, theta=theta)
  reconstruction_fbp = iradon(sinogram, theta=theta, filter_name='ramp')

  # Subplot for Original Image
  plt.subplot(len(projections), 2, 2 * i + 1)
  plt.imshow(image, cmap=plt.cm.Greys_r)  # Display the first image
  plt.axis('off')
  plt.title(f'Original Image')

  # Subplot for Reconstructed Image
  plt.subplot(len(projections), 2, 2 * i + 2)
  plt.imshow(reconstruction_fbp, cmap=plt.cm.Greys_r)
  plt.title(f"Reconstruction\nFiltered back projection using {projection} projections")
  plt.axis('off')

plt.tight_layout()

plt.show()
```

```python
import numpy as np
def shepp_logan_canvas(ellipse = False):
    p1 = np.array([1, .69, .92, 0, 0, 0])
    p2 = np.array([-.98, .6624, .8740, 0, -.0184, 0])
    p3 = np.array([-.2, .1100, .3100, .22, 0, -18])
    p4 = np.array([-.2, .1600, .4100, -.22, 0, 18])
    p5 = np.array([.1, .2100, .2500, 0, .35, 0])
    p6 = np.array([.1, .0460, .0460, 0, .1, 0])
    p7 = np.array([.2, .0460, .0460, 0, -.1, 0])
    p8 = np.array([.1, .0460, .0230, -.08,  -.605, 0])
    p9 = np.array([.1, .0230, .0230, 0, -.606, 0])
    p10 = np.array([.1, .0230, .0460, .06, -.605, 0])
    P = np.vstack((p1, p2, p3, p4, p5, p6, p7, p8, p9, p10))
    if ellipse == True:
      P = np.vstack((p1, p2, p4, p5, p6, p7, p8, p9, p10)) #without p3
    return P

def phantom(canvas, n=300):
	p = np.zeros ((n, n))
	ygrid, xgrid = np.mgrid[1:-1:(1j*n), -1:1:(1j*n)]
	for ellip in canvas:
		I   = ellip [0]
		a2  = ellip [1]**2
		b2  = ellip [2]**2
		x0  = ellip [3]
		y0  = ellip [4]
		phi = ellip [5] * np.pi / 180
		x = xgrid - x0
		y = ygrid - y0
		cos_p = np.cos(phi)
		sin_p = np.sin(phi)
		locs = (((x * cos_p + y * sin_p)**2) / a2
              + ((y * cos_p - x * sin_p)**2) / b2) <= 1
		p [locs] += I
	return p
```

### Bonus 2 : follow the Radon transform of a point with a sinus function

You will try to follow the sinus curve in the sinogram.
Using the link between the cartesian space and the polar space, having (x, y) coordinates you can find the amplitude and phase of the sinus.
Compute the function `polar_transform`

```python
x_new, y_new = 15, 10
size = 40
proj = 100
img = np.zeros((size, size))
img[y_new, x_new] = 1
S = radon(img, theta=np.linspace(0, 360, proj))
offset = size//2
f = 1/proj
proj_vect = np.float64([i for i in range(proj)])

def sinusoidal_curve(x, amplitude, phase):
    return amplitude * np.sin(2 * np.pi * f * x + phase) + offset

def polar_transform(x, y, offset):
    phi = np.arctan2(y, x) + offset
    A = np.sqrt(x**2 + y**2)
    return A, phi

plt.figure(figsize=(30, 5))
plt.subplot(121)
plt.imshow(img, cmap="gray")
plt.plot(x_new, y_new, "*y", label="Point")
plt.plot(offset, offset, "*r", label="Rotation center")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Image of a point at (%.2i, %.2i)" %(x_new, y_new))
plt.legend()

plt.subplot(122)
plt.plot([0, proj_vect[-1]], [offset, offset], "--r", label="Rotation center")
polar_coord = polar_transform(x_new, y_new, offset)
plt.plot(proj_vect, sinusoidal_curve(proj_vect, polar_coord[0], polar_coord[1]), "--", label="Polar coordinates (%.2f, %.2f)" %(polar_coord[0], polar_coord[1]))
plt.imshow(S, cmap="gray")
plt.legend()
plt.title("Sinogram")
plt.xlabel("Projections")
plt.show()
```

```python
from skimage.transform import iradon_sart

reconstruction_sart = np.zeros_like(image)
N_ITERATIONS = 3
snr_list = []

def snr(a, axis=0, ddof=0):
    a = np.asanyarray(a)
    m = a.mean(axis)
    sd = a.std(axis=axis, ddof=ddof)
    return np.where(sd == 0, 0, m/sd)


plt.figure(figsize=(12,8))
for i in range(N_ITERATIONS):
    reconstruction_sart = iradon_sart(sinogram, theta=theta, image=reconstruction_sart)
    error = reconstruction_sart - image

    snr_sart = snr(reconstruction_sart)
    snr_sart_mean = np.sqrt(np.mean(snr_sart**2))
    snr_list.append(snr_sart_mean)

    plt.subplot(1,N_ITERATIONS,i+1)
    im = plt.imshow(reconstruction_sart, cmap=plt.cm.Greys_r)
    cbar = plt.colorbar(im)
    cbar.set_label('Intensity')
    plt.title(f'SART \nIteration : {i + 1}\nReconstruction error : {np.sqrt(np.mean(error**2)):.3g}\nSNR : {snr_sart_mean:.3g}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

The Simultaneous Algebraic Reconstruction Technique, is an iterative method used for reconstructing an image from a set of projections, meaning that it starts with an initial estimate of the image and then refines this estimate through multiple iterations until convergence is achieved.
<br>
In each iteration of SART, it computes a correction to the current image estimate based on the discrepancy between the measured data and the data predicted by the current estimate. It updates the image estimate incrementally and repeats this process until a satisfactory reconstruction is achieved.

We will work with a 3D. Tag Image File Format (TIFF) file a format widely used in tomography.

Digital Imaging and Communications in Medicine (DICOM) is the standard for the communication and management of medical imaging information and related data.
In python, we will use the library `tiffile` to load the image.
To work with 3D interactively, `plotly` is a good librairy.

```python
import tifffile
import plotly.graph_objects as go
# function to convert transmission based image into attenuation images
def correction_norm(matrix):
    matrix = -np.log(matrix)
    mn = np.min(matrix)
    matrix_zero_bound = matrix - mn

    mx = np.max(matrix_zero_bound)
    matrix_one_bound = matrix_zero_bound/mx
    return matrix_one_bound

# recreate the 3D volume
def compute_proj_correction(matrix):
    matrix = np.uint16(matrix)
    proj_corrected_3D = np.empty((matrix.shape[0], matrix.shape[1], matrix.shape[2]))
    for i in range(matrix.shape[0]):
        c1 = correction_norm(matrix[i])
        proj_corrected_3D[i] = c1
    return proj_corrected_3D

# get the sinogram
def _3D_to_sinogramme_corr(path, num):
    obj_3D = tifffile.imread(path)
    obj_3D_sino = np.transpose(obj_3D, (1, 2, 0))
    obj_3D_sino = obj_3D_sino[:, 1:-1, :]
    obj_3D_sino_corr = compute_proj_correction(obj_3D_sino)
    obj_2D_sino_corr = obj_3D_sino_corr[num]
    plt.figure(figsize=(10,10))
    plt.imshow(obj_2D_sino_corr, cmap="gray")
    plt.title("Sinogram at the height of %.i" %num)
    plt.xlabel("Projections")
    plt.colorbar()
    plt.show()
    return obj_2D_sino_corr

# get the 3D plot
def plot_3D(title, matrix, proj, size):
  x = np.array([i for i in range(proj)])
  y = np.array([i for i in range(size)])
  X, Y = np.meshgrid(x, y)
  Z = matrix
  fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
  fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
  fig.update_layout(title=title,
                  width=1000, height=1000,
                  margin=dict(l=65, r=50, b=65, t=90),
                    scene=dict(
            xaxis_title='Projections',
            yaxis_title='Image size',
            zaxis_title='Intensities'
        ))
  fig.show()
```
