## DICOM:  Digital Imaging and Communications in Medicine standard

It is just an image format that is used for viewing different medical images acquired by different modalities. DICOM differs from other image formats in that it groups information into data sets. A DICOM file consists of a header and image data sets, all packed into a single file.






Ref: https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3354356/

**METHOD 1**: Naive Method
Divide the pixels values by **MAXVAL** and display this image

- Hint check the use of `np.clip` to keep pixels that are within a range to set the rest to 0

**MAXVAL** = 255

```python
MAXVAL = 255
START_POINT = (100, 200)
END_POINT = (450, 200)
```

```python
def obtain_intensity_profile(image, start_point, end_point):
    #Getting coordinates of start and end points
    x1, y1 = start_point
    x2, y2 = end_point
    
    line_pixels = np.linspace(start_point, end_point, num=int(np.sqrt((x2 - x1)**2 + (y2 - y1)**2)))
    
    intensity_profile = [image[int(y), int(x)] for x, y in line_pixels]
    return intensity_profile
```

```python
def normalization(image):
    normalized_image = image / MAXVAL

    normalized_image = np.clip(normalized_image, 0, MAXVAL)

    print(f'Minimum Value after normalization: {int(np.min(normalized_image))}')
    print(f'Maximum Value after normalization: {int(np.max(normalized_image))}')

    image_profile = obtain_intensity_profile(image, START_POINT, END_POINT)
    normalized_image_profile = obtain_intensity_profile(normalized_image, START_POINT, END_POINT)

    plt.figure(figsize=(15, 13))

    #Plot of Original Image
    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap="gray")
    plt.axis('off')
    plt.title(f'Original Brain MRI')
    plt.colorbar()

    #Plot of normalized Image
    plt.subplot(2, 2, 2)
    plt.imshow(normalized_image, cmap="gray")
    plt.axis('off')
    plt.title(f'Normalized Brain MRI')
    plt.colorbar()

    plt.subplot(2, 2, 3)
    plt.plot(image_profile, label='Intensity Profile of original Image', color='blue', linewidth=2)
    plt.plot(normalized_image_profile, label='Intensity Profile of Normalized Image', color='red', linewidth=2)
    plt.legend()

    plt.show()
    
normalization(image)
```

**ANSWER**

It can be observed from the plotted intensity profiles that the range of pixel (intensity) values of the original image frange from 0 to as far as 1200, while the normalized image ranges from 0 to 5, and since both of them were plotted on the same scale, the original image's intensity profile is more informative as its values are more suited to the scale, so slight drops in the pixel values can be observed, which cannot be done for the normalized image as the min and max values are so small that both the max values and  changes are not reflected in the plot.

### **METHOD 2**: Min-Max Method. 

Find the min and max value of the pixels. Use these values to set the range of the pixels to be in [0, **MAXVAL**]

Use this and display the new image that you get

```python
#Implementing min-max normalization
def min_max_normalization(image):
    min_maxed_image = (image - np.min(image)) / (np.max(image) - np.min(image)) * MAXVAL
    min_maxed_image = np.clip(min_maxed_image, 0, MAXVAL).astype(np.uint8)

    image_profile = obtain_intensity_profile(image, START_POINT, END_POINT)
    min_maxed_image_profile = obtain_intensity_profile(min_maxed_image, START_POINT, END_POINT)

    plt.figure(figsize=(15, 13))

    plt.subplot(2, 2, 1)
    plt.imshow(image, cmap="gray")
    # plt.axis('off')
    plt.title(f'Brain MRI')
    plt.colorbar()

    #Plot of min-maxed Image
    plt.subplot(2, 2, 2)
    plt.imshow(min_maxed_image, cmap="gray")
    plt.axis('off')
    plt.title(f'Min-Maxed Brain MRI')
    plt.colorbar()

    plt.subplot(2, 2, 3)
    plt.plot(image_profile, label='Intensity Profile of original Image', color='blue', linewidth=2)
    plt.plot(min_maxed_image_profile, label='Intensity Profile of Min-Maxed Image', color='red', linewidth=2)
    plt.legend()

    plt.show()

min_max_normalization(image)
```

**ANSWER**

It can be observed from the plotted intensity profiles that the range of pixel (intensity) values of the original image frange from 0 to as far as 1200, while the min-max normalized image ranges from 0 to 255, and since both of them were plotted on the same scale, the original image's intensity profile is more informative as its values are more suited to the scale, but this differs from the previously normalized image as the min-max normalized range is a bit higher (0-255) compared to (0-5), so at least significant changes in pixel values can be observed in the plot, with slight changes in the pixel values being less observable.
