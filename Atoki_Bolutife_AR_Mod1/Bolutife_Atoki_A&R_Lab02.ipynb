```python
from skimage.data import shepp_logan_phantom
from skimage.transform import radon, iradon, rescale, resize
from skimage.transform.radon_transform import _get_fourier_filter
from skimage.transform import iradon_sart

from scipy.fft import fft, ifft, fftfreq, fftshift
from scipy import ndimage

import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
IM_SIZE = 100
N_PROJECTIONS = 200
ANGLE = 360
```

```python
def radon_scratch(image, theta):
    size = image.shape[0]
    sinogram_pix = np.zeros((size, len(theta)), dtype=np.float16)
    rot_pix = np.zeros((size, size))
    center = size//2
    for i, angle in enumerate(theta):
        c, s = np.cos(np.deg2rad(angle)), np.sin(np.deg2rad(angle))
        R = np.array([[c, s], [-s, c]])
        for j in range(size):
            for k in range(size):
                x, y = np.array([j, k]) - center
                x_rot, y_rot = np.dot(R, np.array([x, y]))
                x_rot_idx, y_rot_idx = np.uint8(x_rot + center), np.uint8((y_rot + center))
                if x_rot_idx >= 0 and x_rot_idx < size and y_rot_idx >= 0 and y_rot_idx < size:
                    rot_pix[x_rot_idx, y_rot_idx] = image[j, k]
                sinogram_pix[:, i] = rot_pix.sum(axis=0)
    return sinogram_pix
```

```python
image = shepp_logan_phantom()
image = rescale(image, scale=0.4, mode='reflect', channel_axis=None)
image = resize(image, (IM_SIZE, IM_SIZE))

theta = np.linspace(0., ANGLE, N_PROJECTIONS, endpoint=False)

start_time = time.time()
sinogram_from_scratch = radon_scratch(image, theta)
scratch_time = time.time() - start_time

start_time = time.time()
sinogram_from_skimage = radon(image, theta=theta)
sk_time = time.time() - start_time

sinograms = [sinogram_from_scratch, sinogram_from_skimage]
titles = ['Sinogram from scratch', 'sinogram from skimage']
times = [scratch_time, sk_time]

plt.figure(figsize=(14, 8))
for i in range(2):
  dx, dy = 0.5 * ANGLE / max(image.shape), 0.5 / sinograms[i].shape[0]

  # Subplot for Original Image
  plt.subplot(2, 2, 2 * i + 1)
  plt.imshow(image, cmap=plt.cm.Greys_r)  # Display the first image
  plt.axis('off')
  plt.title(f'Original Image')

  # Subplot for Sinogram
  plt.subplot(2, 2, 2 * i + 2)
  plt.imshow(sinograms[i], cmap=plt.cm.Greys_r,
           extent=(-dx, ANGLE + dx, -dy, sinograms[i].shape[0] + dy),
           aspect='auto')
  plt.xlabel("Projection angle (deg)")
  plt.ylabel("Projection position (pixels)")
  plt.title(f'Radon transform\n ({titles[i]}) \n computed in {times[i]} seconds')

plt.tight_layout()

plt.show()
```

### Inverse Radon function

- Keep the sinogram generated by the `radon` function
- You will implement the Filtered Back Projection algorithm
- First, run the inverse Radon transform.
- Indeed, we won't filter the sinogram

#### Sources

- FBP reconstruction video : https://www.youtube.com/watch?v=pZ7JlXagT0w&ab_channel=ASTRAToolbox
- Help to write the algorithm : https://howradiologyworks.com/filtered-backprojection-fbp-illustrated-guide-for-radiologic-technologists/

```python
def inverse_radon(sinogram, theta):
    reconstruction = np.zeros((SIZE, SIZE), dtype=np.float64)
    stack = reconstruction
    for i, angle in enumerate(theta):
            proj_vect = sinogram[:, i]
            mat_proj = np.tile(proj_vect, (SIZE, 1))
            rot_pix = ndimage.rotate(mat_proj, angle, reshape=False)
            reconstruction = rot_pix
            stack = stack + reconstruction
    return stack
```

```python
def inverse_radon_norm_modified(sinogram, theta, filter_name):

    sinogram_shape = sinogram.shape[0]
    projection_size_padded = max(64, int(2 ** np.ceil(np.log2(2 * sinogram_shape))))
    pad_width = ((0, projection_size_padded - sinogram_shape), (0, 0))
    img = np.pad(sinogram, pad_width, mode='constant', constant_values=0)

    fourier_filter = _get_fourier_filter(projection_size_padded, filter_name)
    projection = fft(img, axis=0) * fourier_filter
    radon_filtered = np.real(ifft(projection, axis=0)[:sinogram_shape, :])

    reconstruction = np.zeros((SIZE, SIZE), dtype=np.float64)
    stack = reconstruction
    for i, angle in enumerate(theta):
            proj_vect = radon_filtered[:, i]
            mat_proj = np.tile(proj_vect, (SIZE, 1))
            rot_pix = ndimage.rotate(mat_proj, angle, reshape=False)
            reconstruction = rot_pix
            stack = stack + reconstruction
    return stack * np.pi / (2 * len(theta))


```

### 1D approach

- The algebraic reconstruction technique (ART) is an iterative reconstruction technique used in computed tomography. It reconstructs an image from a series of angular projections (a sinogram).

- First we will understand the update of the coefficients in one dimension.
- Go to the https://scipython.com/blog/visualizing-kaczmarzs-algorithm/
- Copy and paste the code
- What do you understand on the update of coefficients ?
- Iterativelly, we can solve a square matric which respects the invertibility theorem

```python
# code from https://scipython.com/blog/visualizing-kaczmarzs-algorithm/
A = np.array([[-4, 1], [2, 0.5], [3, 1.5], [0, 1]])
b = np.array([-2, 3, 6, 2.])
m, n = A.shape
A2 = np.array([[-4, 1], [2, 0.5], [3, 1.5], [0, 1]])
b2 = np.array([-2, 3, 4, 2.])
def solve_kaczmarz(A, b, TOL=1.e-6, randomize=False):
    def calc_err(X):
        return np.max(np.abs(A @ X - b))
    X = np.zeros(n)
    Xhistory = [X]
    err_history = [calc_err(X)]
    k = 0
    if randomize:
        p = (np.linalg.norm(A, axis=1) / np.linalg.norm(A))**2
    else:
        p = range(m)
    while True and k < 20:
        if randomize:
            i = np.random.choice(range(m), p=p)
        else:
            i = k % m
        ai = A[i,:]
        Xnew = X + (b[i] - ai @ X) / np.linalg.norm(ai)**2 * ai
        err = np.max(np.abs(A @ Xnew - b))
        Xhistory.append(Xnew)
        err_history.append(err)
        if err < TOL:
            break
        X = Xnew
        k += 1
    return X, np.array(Xhistory), np.array(err_history)

def plot_trajectory(Xhistory, A, b):
    x = np.linspace(-5, 5, 2)
    M = np.vstack((x, np.ones(2))).T
    for i in range(m):
        y = M @ (-A[i,0], b[i]) / A[i,1]
        plt.plot(x, y)
    plt.plot(Xhistory[:,0], Xhistory[:,1], 'k-o')
    plt.axis('square')
    plt.xlim(-0.5, 2.5)
    plt.ylim(-0.5, 2.5)
    plt.xlabel('x')
    plt.ylabel('y')

plt.figure(figsize=(30, 5))
plt.subplot(131)
X, Xhistory, _ = solve_kaczmarz(A, b, 1.e-2, randomize=False)
plot_trajectory(Xhistory, A, b)
print(f'Solution vector: {X}.')
plt.subplot(132)
X, Xhistory, _ = solve_kaczmarz(A, b, 1.e-2, randomize=True)
plot_trajectory(Xhistory, A, b)
print(f'Solution vector: {X}.')
plt.subplot(133)
X, Xhistory, _ = solve_kaczmarz(A2, b2, 1.e-2, randomize=False)
plot_trajectory(Xhistory, A2, b2)
print(f'Solution vector: {X}.')
plt.show()
```

### ART update one batch

- Use the ART function to update pixel of a random matrix
- Go through all the sinogram only one time
- Measure the execution time, compare with FBP and SART from skimage, and evaluate the results

#### Sources
- Introduction video : https://www.youtube.com/watch?v=hkJHL0DwUUQ&ab_channel=ASTRAToolbox
- Update pixels of the image : https://en.m.wikipedia.org/wiki/Algebraic_reconstruction_technique

```python
def ART_one_batch(sinogram, theta):
  size = sinogram.shape[0]
  X_rot = np.zeros((size, size))
  for a, angle in enumerate(theta):
    X_rot = ndimage.rotate(X_rot, - angle, reshape=False)
    S_calc = np.sum(X_rot, axis=0)
    diff = sinogram[:,a] - S_calc
    norm = np.linalg.norm(sinogram[:,a])

    for i in range(size):
        for j in range(size):
            X_rot[i,j] = X_rot[i,j] + diff[j]/norm
            if X_rot[i,j] < 0.01:
              X_rot[i,j] = 0
    X_rot = ndimage.rotate(X_rot, angle, reshape=False)
  return X_rot
```

### ART

- Modify the function to have multiple iterations with a stop condition on the plateau
- Use the criterion of your choice for one of the stopping condition
- The number of iterations would be the other stop condition.
- Plot the profile at the end of each global iteration compared to the one with reconstruction with skimage

```python
def ART_multiple_iter(sinogram, theta, max_iterations, plateau_stop=1e-5):
    size = sinogram.shape[0]
    X_rot = np.zeros((size, size))
    profile_history = []

    for _ in range(max_iterations):
        init_X_rot = X_rot.copy()

        for a, angle in enumerate(theta):
            X_rot = ndimage.rotate(X_rot, -angle, reshape=False)
            S_calc = np.sum(X_rot, axis=0)
            diff = sinogram[:, a] - S_calc
            norm = np.linalg.norm(sinogram[:, a])

            for i in range(size):
                for j in range(size):
                    X_rot[i, j] = X_rot[i, j] + diff[j] / norm
                    if X_rot[i, j] < 0.01:
                        X_rot[i, j] = 0

            X_rot = ndimage.rotate(X_rot, angle, reshape=False)

        profile_history.append(np.sum(X_rot, axis=0))

        change = np.sum(np.abs(X_rot - init_X_rot))

        if change < plateau_stop:
            break

    return X_rot, profile_history
```

```python
max_iters = 10
reconstructed_image, profile_history = ART_multiple_iter(sinogram_from_skimage, theta, max_iters)
error_art_new = reconstructed_image - image
print(f'Error = {np.sqrt(np.mean(error_art_new**2)):.3g}')

# Plot profiles at the end of each iteration
plt.figure(figsize=(12, 6))
plt.title('Profiles Comparison')
plt.xlabel('Projection Angle')
plt.ylabel('Intensity')

# Reconstruct using skimage for comparison
reconstruction_fbp = iradon(sinogram_from_skimage, theta=theta, filter_name='ramp')
plt.plot(reconstruction_fbp.sum(axis=0), label='Library Function', linestyle='--')

for i, profile in enumerate(profile_history):
    plt.plot(profile, label=f'Iteration {i+1}')

plt.legend()
plt.show()
```

The function `ART_multiple_iter` takes several input parameters, including the sinogram data (`sinogram`), a list of projection angles (`theta`), the maximum number of iterations (`max_iterations`), and a stopping criterion (`plateau_stop`) based on the change in the reconstructed image profile. The algorithm checks for convergence by monitoring the change in the image profile (`change`). If the change falls below the specified `plateau_stop`, the reconstruction stops before reaching the maximum number of iterations.
